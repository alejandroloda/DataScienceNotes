## Transformar etiquetas en números

### Carga de datos

```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

iris = load_iris()
print(iris.keys())

df = pd.DataFrame(data=np.c_[iris['data'], iris['target']],
                  columns=iris['feature_names'] + ['target'])
iris_names={0:'Iris-setosa',1:'Iris-versicolor',2:'Iris-virginica'}
df.target = df.target.apply(lambda val:iris_names[int(val)])
df
```

### Uso

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(['Iris-setosa','Iris-versicolor','Iris-virginica'])
df['target'] = le.transform(df['target'])
df
```
